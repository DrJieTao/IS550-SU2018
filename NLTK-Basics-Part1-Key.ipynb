### 1.3 Statistics of Text

How can we automatically identify the words of a text that are most informative about the topic and genre of the text? Imagine how you might go about finding the 50 most frequent words of a book. One method would be to keep a tally for each vocabulary item, like that shown in below. The tally would need thousands of rows, and it would be an exceedingly laborious process  so laborious that we would rather assign the task to a machine.

<img src = 'http://www.nltk.org/images/tally.png' alt = "Word Tally" align = 'center' style = "width: 25%; height: 25%"/>

The table above is known as a <b>frequency distribution</b>, and it tells us the frequency of each vocabulary item in the text. (In general, it could count any kind of observable event.) It is a "distribution" because it tells us how the total number of word tokens in the text are distributed across the vocabulary items. Since we often need frequency distributions in language processing, NLTK provides built-in support for them. Let's use a *FreqDist* to find the 50 most frequent words of *Moby Dick*:

```python
fdist1 = FreqDist(text1)
print fdist1
print fdist1.most_common(50)
```

```python
# Now let us look up a certain word "whale" - find out the frequency of it
fdist1['whale']
```

```python
# we can also plot the freqencies of the 50 words
fdist1.plot(50, cumulative = 1)
```

#### Fine-grained Selection of Words

Next, let's look at the long words of a text; perhaps these will be more characteristic and informative. For this we adapt some notation from set theory. We would like to find the words from the vocabulary of the text that are more than 15 characters long. <BR>
Let's call this property *P*, so that *P(w)* is true if and only if w is more than 15 characters long. Now we can express the words of interest using mathematical set notation as shown below. This means "the set of all w such that w is an element of V (the vocabulary) and w has property P".

In set theory: {w | w  V & P(w)} <BR>
In programming language: [w for w in V if p(w)]

```python
# so how to implement this in Python?
## Get unique words in the disctionary
V = set(text1)
## select unique words from dictionary when length is greater than 15
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

#### Collocations and Bigrams

A <B>collocation</B> is a sequence of words that occur together unusually often. Thus *red wine* is a collocation, whereas *the wine* is not. <BR>
A characteristic of collocations is that they are resistant to substitution with words that have similar senses; for example, *maroon wine* sounds definitely odd.<BR>
To get a handle on collocations, we start off by extracting from a text a list of word pairs, also known as <b>bigrams</b>. This is easily accomplished with the function *bigrams()*:

```python
from nltk import bigrams
list(bigrams(['more', 'is', 'said', 'than', 'done']))
```

Here we see that the pair of words than-done is a bigram, and we write it in Python as ('than', 'done'). <BR> Now, collocations are essentially just frequent bigrams, except that we want to pay more attention to the cases that involve rare words. <BR>In particular, we want to find bigrams that occur more often than we would expect based on the frequency of the individual words. The *collocations()* function does this for us. We will see how it works later.

```python
text4.collocations()
```

```python
text8.collocations()
```
